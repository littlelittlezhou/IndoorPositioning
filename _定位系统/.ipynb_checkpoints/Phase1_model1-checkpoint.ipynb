```python
import IPYNB_IMPORTER
import Phase2
from DivisionForP2 import Dataset
import numpy as np
import keras
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
from keras.utils import to_categorical
from keras.models import Sequential, load_model
from keras.layers import Dense, Activation,Dropout,Conv2D,MaxPooling2D, Flatten
from keras.optimizers import RMSprop,Nadam,Adam
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
import random
```

```python
class LossHistory(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.losses = {'batch':[], 'epoch':[]}
        self.accuracy = {'batch':[], 'epoch':[]}
        self.val_loss = {'batch':[], 'epoch':[]}
        self.val_acc = {'batch':[], 'epoch':[]} 
        
    def on_batch_end(self, batch, logs={}):

        self.losses['batch'].append(logs.get('loss'))
        self.accuracy['batch'].append(logs.get('accuracy'))
        self.val_loss['batch'].append(logs.get('val_loss'))
        self.val_acc['batch'].append(logs.get('val_accuracy'))
        
    def on_epoch_end(self, batch, logs={}):
        self.losses['epoch'].append(logs.get('loss'))
        self.accuracy['epoch'].append(logs.get('accuracy'))
        self.val_loss['epoch'].append(logs.get('val_loss'))
        self.val_acc['epoch'].append(logs.get('val_accuracy'))

    def loss_plot(self, loss_type,modelname):
        iters = range(len(self.losses[loss_type]))
    
        # acc
        plt.figure()
        plt.plot(iters, self.accuracy[loss_type], 'r', label='train acc')
        if loss_type == 'epoch':
            # val_acc
            plt.plot(iters, self.val_acc[loss_type], 'b', label='val acc')
        plt.grid(True)
        plt.xlabel(loss_type)
        plt.ylabel('acc-loss')
        plt.legend(loc="upper right")
        plt.savefig(modelname+"_acc.png")
        plt.show()

        # loss
        plt.figure()
        plt.plot(iters, self.losses[loss_type], 'g', label='train loss')
        if loss_type == 'epoch':
            # val_loss
            plt.plot(iters, self.val_loss[loss_type], 'k', label='val loss')
        plt.grid(True)
        plt.xlabel(loss_type)
        plt.ylabel('acc-loss')
        plt.legend(loc="upper right")
        plt.savefig(modelname+"_los.png")
        plt.show()
```

```python

def datagen(X,y,normalizer,batch_size=32):
    # 第一步：计算各个区域的每AP的属性。
    # 属性：为0的概率p。
    #       取值历史最小值min。
    #       取值历史最大值max。
    x_div_by_area,y_div_by_area = Phase2.DivByArea(X,y)
    attr = {}
    for area,samples in x_div_by_area.items():          #对于每一个区域
        total = len(samples)    # 该区域样本总个数
        zero = sum(samples == 0)  # 该区域每一列0的个数
        ratio = zero/total       # 该区域样本每个ap取0的可能性
        min_ = np.amin(samples,axis=0)   # 该区域可以取到的最小值
        samples[samples==0] = -120
        max_ = np.amax(samples,axis=0)   # 该区域可以取到的最大值
        # 已计算出 ratio min_ max_
        attr[area] = (ratio,min_,max_)
    while True:  
        vXs = []
        vys = []       
        for i in range(batch_size):
            area = random.randint(1,32)     # 本次随机的样本所处区域
            vX = []                # 预设一个虚拟X
            #取出需要的区域参数
            ratio = attr[area][0]
            min_ = attr[area][1]
            max_ = attr[area][2]
            for apnum in range(len(X[0])):
                if random.random() < ratio[apnum]:    # 随机取样，如果在取0 的比例中，为0 
                    vX.append(0)
                else:         # 随机结果认为该ap取值不应为0
                    temp = random.randint(min_[apnum],max_[apnum])
                    error = random.randint(-3,3)
                    temp = temp+error
                    vX.append(temp)
            vXs.append(vX)
            vys.append(area)
            
        ###################### 下面是预处理流程
        vXs = np.array(vXs)   # 转为numpy数组
        vys = np.array(vys) 
        vXs[vXs==0] = -120    # 0赋-120
        vXs = normalizer.transform(vXs)    # 归一
        
        vys = vys-1
        vys = to_categorical(vys,32)    
        yield (vXs,vys)
```

```python
y_test,X_test = Dataset("Test_0.csv")
y_train,X_train = Dataset("Train_0.csv")

X_temp = np.concatenate((X_train,X_test),axis=0)
X_temp[X_temp==0] = -120
normalize = MinMaxScaler()
normalize.fit(X_temp)

X_test[X_test==0] = -120
X_test = normalize.transform(X_test)

area_test = y_test[:,0]
area_test = area_test -1
area_test = to_categorical(area_test,32) 
area_true = np.argmax(area_test,axis=1)+1
```

```python
# model_1, 全连接神经网络，形状 469（输入）-400-100-32（输出）
# 其他细节：activation：relu，softmax（输出）
# optimizer：rmsprop（lr=0.001）
# epochs:100,batch_size:32

model_1 = Sequential([
    Dense(400, input_dim=469),
    Activation('relu'),
    Dense(100),
    Activation('relu'),
    Dense(32),
    Activation('softmax'),
])

# 定义优化器
rmsprop = RMSprop(lr=0.001, rho=0.9, epsilon=1e-08, decay=0.0)
model_1.compile(optimizer=rmsprop,
              loss='categorical_crossentropy',
              metrics=['accuracy']) # metrics赋值为'accuracy'，会在训练过程中输出正确率

history = LossHistory()

print('Training ------------')
#model_1.fit(X_train, area_train, validation_data=(X_test,area_test), epochs=50, batch_size=32, callbacks=[history])
model_1.fit_generator(datagen(X_train,y_train,normalize), steps_per_epoch = 60, epochs = 50, validation_data=(X_test,area_test), callbacks=[history])
print('\nTesting ------------')
# 评价训练出的网络
loss, accuracy = model_1.evaluate(X_test, area_test)

print('test loss: ', loss)
print('test accuracy: ', accuracy)
history.loss_plot('epoch',"./Phase_1/model_1_0")

model_1.save("./Phase_1/model_1_0.h5")
```
