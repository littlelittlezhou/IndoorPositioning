```python
# model_2 试图复现使用rssi相似度的定位方法
# rssi相似度定义为s = sqrt（Σ（rssi-rssi0）^2）
# 具体参见论文
from DivisionForP2 import Dataset
from sklearn.utils import shuffle
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
def model_2_pred(test_x,train_x,area_ref):
    # 上面用于得到用于比对的训练集，而不需要在参数位置输入
    # 下面是函数主体
    areas = []
    for i in range(len(test_x)):
        S = []
        for j in range(len(train_x)):
            diff = train_x[j]-test_x[i]
            Si = np.sqrt(np.sum(diff**2))
            S.append(Si)
        area = area_ref[np.argmin(S)]
        areas.append(area)
    return areas
```

```python
if __name__=='__main__':
    labels,samples = Dataset("Vec_Del_dataset_0.csv")
    x,y = shuffle(samples,labels)   # 打乱数据
    X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
    pred_area = model_2_pred(X_test,X_train,y_train[:,0])
    area_true = y_test[:,0]
    judge = pred_area==area_true
    ratio = np.sum(judge!=0)/len(area_true)
    print("Area correct ratio:",ratio)
    for index,pred in enumerate(pred_area):
        if pred != area_true[index]:
            print("area wrong:(ture,pred)=",area_true[index],pred_area[index])
```
