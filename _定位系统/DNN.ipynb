```python
import numpy as np
import pandas as pd
import keras
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
from keras.utils import to_categorical
from keras.models import Sequential, load_model
from keras.layers import Dense, Activation,Dropout,Conv2D,MaxPooling2D, Flatten
from keras import optimizers
from keras import regularizers
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from keras import models
from keras import layers
import random
import csv
from sklearn.model_selection import KFold
```

```python
class LossHistory(keras.callbacks.Callback): # 回调函数, 用于查看训练模型的内在状态和统计
    #  在训练时， 相应的回调函数的方法会被在各自的阶段被调用
    def on_train_begin(self, logs={}): # 在模型开始训练时被调用
        self.losses = {'batch':[], 'epoch':[]}              #创建每批结束和每轮结束的记录数组
        self.accuracy = {'batch':[], 'epoch':[]}           
        self.val_loss = {'batch':[], 'epoch':[]}
        self.val_accuracy = {'batch':[], 'epoch':[]} 
        
    def on_batch_end(self, batch, logs={}): # 在每批结束时被调用

        self.losses['batch'].append(logs.get('loss'))       #记录每批结束的训练损失
        self.accuracy['batch'].append(logs.get('accuracy')) #记录每批结束的训练精确度
        self.val_loss['batch'].append(logs.get('val_loss')) #记录每批结束的验证损失
        self.val_accuracy['batch'].append(logs.get('val_accuracy'))   #记录每批结束的验证精确度 
        
    def on_epoch_end(self, batch, logs={}): # 在每轮结束时被调用
        self.losses['epoch'].append(logs.get('loss'))       #记录每轮结束的训练损失
        self.accuracy['epoch'].append(logs.get('accuracy')) #记录每轮结束的训练精确度
        self.val_loss['epoch'].append(logs.get('val_loss')) #记录每轮结束的验证损失
        self.val_accuracy['epoch'].append(logs.get('val_accuracy'))   #记录每轮结束的验证精确度

    def loss_plot(self, loss_type,modelname):
        iters = range(len(self.losses[loss_type]))
    
        # acc loss
        plt.figure()
        plt.plot(iters, self.accuracy[loss_type], 'r', label='train accuracy')
        
        plt.plot(iters, self.losses[loss_type], 'g', label='train loss')
        plt.plot(iters, self.val_accuracy[loss_type],'yo',label='Validation accuracy')
        plt.plot(iters, self.val_loss[loss_type],'bo',label='Validation loss')
        plt.grid(True)
        plt.xlabel(loss_type)
        plt.ylabel('acc-loss')
        plt.legend(loc="upper right")
        plt.savefig(modelname+"_acc_{}.png".format(k)) #保存为图片
```

```python
# 导入数据
#该方法用于引入数据集，数据集要求为已经向量化的数据
#输入为文件名，默认在/Data文件夹中，数据形式为【区域，x，y，样本】
#返回为labels，samples，同索引的对应
def Dataset(Filename):
    datas = pd.read_csv("./Data/"+Filename,header=None)
    datas = np.array(datas)
    labels = datas[:,:3]    # csv数据中，前三列为label：区域，x，y
    samples = datas[:,3:]   # 之后的为样本
    return labels,samples
```

```python
# 定义model_1

def model_1():
    model = models.Sequential()
    model.add(layers.Dense(300,activation='relu',input_dim=469))
#     model.add(layers.Dropout(0.5))
    model.add(layers.Dense(200,kernel_regularizer=keras.regularizers.l2(0.01),activation='relu'))
    model.add(layers.Dropout(0.5))
    model.add(layers.Dense(64,activation='relu'))
    model.add(layers.Dense(32,activation='softmax'))
    
#     rmsprop = RMSprop(lr=0.001, rho=0.9, epsilon=1e-08, decay=0.0)
    adam = optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
    model.compile(loss='categorical_crossentropy', optimizer=adam,metrics=['accuracy'])
#     model.compile(optimizer=rmsprop,
#                   loss='categorical_crossentropy',
#                   metrics=['accuracy'])
    return model
```

```python
# K折验证数据划分

all_histories = [] # 存储所有的history
model_1 = model_1() # 构建模型
model_1.summary() #总结模型


# test_y,test_x = Dataset("Test_0.csv")
# train_y,train_x = Dataset("Train_0.csv")

test_y,test_x = Dataset("Test_1.csv")
train_y,train_x = Dataset("Train_1.csv")
# 对标签进行处理
test_y = test_y[:,0] # 区域号位于第0列，所以取出区域号
test_y -= 1 # 将原本区域号是1~32，对应到0~31
y_test = to_categorical(test_y,32) # one-hot编码
# 对标签进行处理
train_y = train_y[:,0] # 区域号位于第0列，所以取出区域号
train_y -= 1 # 将原本区域号是1~32，对应到0~31
y_train = to_categorical(train_y,32) # one-hot编码

# 对样本数据进行处理
test_x[test_x == 0] = -120 # 将信号强度为0的数据，设置为-120 表示Rssi最弱
normalize = MinMaxScaler() # 引入归一化方法，即 按axis=0，（当前值-最小值）/（最大值-最小值）
X_test = normalize.fit_transform(test_x) # fit_transform 会先拟合部分数据，找到normalize中所要用到的特征值
# 对样本数据进行处理
train_x[train_x == 0] = -120 # 将信号强度为0的数据，设置为-120 表示Rssi最弱
normalize = MinMaxScaler() # 引入归一化方法，即 按axis=0，（当前值-最小值）/（最大值-最小值）
X_train = normalize.fit_transform(train_x) # fit_transform 会先拟合部分数据，找到normalize中所要用到的特征值


# 划分测试集
# X_train,X_test,y_train,y_test = train_test_split(samples_normalized,labels,test_size=0.2) #划分20%的数据作为测试数据
#  K折交叉验证
kf = KFold(n_splits=5, shuffle=True) # 5折交叉验证
k = 0 # 用于记录验证次数,初始化为0
for tra_index, val_index in kf.split(X_train,y_train):
#     print(" Train Index:\n", tra_index)
    X_train_tra = X_train[tra_index]
    print(X_train_tra.shape)
    X_train_val = X_train[val_index]
    print(X_train_val.shape)
#     print(" Val Index:\n", val_index)
    y_train_tra = y_train[tra_index]
    y_train_val = y_train[val_index]
    print(y_train_tra.shape)
    print(y_train_val.shape)
    k += 1 # 验证次数+1
    history = LossHistory()
    print('Training 第{}次 ------------'.format(k))
    model_1.fit(X_train_tra, y_train_tra, batch_size=32,epochs=50,  callbacks=[history], validation_data=(X_train_val,y_train_val))
    print('\nTesting 第{}次 ------------'.format(k))
    # 评价训练出的网络
    loss, accuracy = model_1.evaluate(X_test, y_test)
    print('test loss: ', loss)
    print('test accuracy: ', accuracy)
    history.loss_plot('epoch',"./Phase_1/model_1")
    all_histories.append(history) #加入每次的记录
    
model_1.save("./Phase_1/model_1.h5")
# model_1.save("./Phase_1/model_0.h5")
```
